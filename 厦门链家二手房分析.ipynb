```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
from IPython.display import display 
plt.style.use("fivethirtyeight") 
sns.set_style({'font.sans-serif':['simhei','Arial']}) 
%matplotlib inline

city='厦门'
with sqlite3.connect('DetailInfo.db') as conn:
    df=pd.read_sql_query('SELECT * FROM %s' % city,conn)
```

```python
new_columns=['roomNum', 'square','buildingArea', 'buildYear', 'isNew', 'ctime', 'mtime', 'orientation',
       'floorStat', 'totalFloor', 'decorateType', 'hbtName', 'isYezhuComment',
       'isGarage', 'houseType', 'isFocus', 'status', 'isValid','address', 
       'communityName','districtName','regionName', 'subwayInfo',
       'schoolName', 'bizcircleFullSpell','price', 'unitPrice', 'listPrice', 'publishTime', 'isVilla',
       'villaNoFloorLevel', 'villaName']
df1=pd.DataFrame(df,columns=new_columns)
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1['region']=df1['bizcircleFullSpell']
```

```python
df1.head()
```

```python
# 对二手房区域分组对比二手房数量和每平米房价 
df_house_count = df1.groupby('region')['price'].count().sort_values(ascending=False).to_frame().reset_index() 
df_house_mean = df1.groupby('region')['unitPrice'].mean().sort_values(ascending=False).to_frame().reset_index() 

f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(20,15)) 
sns.barplot(x='region', y='unitPrice', palette="Blues_d", data=df_house_mean, ax=ax1) 
ax1.set_title('各区二手房每平米单价对比',fontsize=15) 
ax1.set_xlabel('区域') 
ax1.set_ylabel('每平米单价') 

sns.barplot(x='region', y='price', palette="Greens_d", data=df_house_count, ax=ax2) 
ax2.set_title('各区二手房数量对比',fontsize=15) 
ax2.set_xlabel('区域') 
ax2.set_ylabel('数量') 

sns.boxplot(x='region', y='price', data=df1, ax=ax3) 
ax3.set_title('各区二手房房屋总价',fontsize=15) 
ax3.set_xlabel('区域') 
ax3.set_ylabel('房屋总价') 
plt.show()
```

```python
#Size特征分析
f, [ax1,ax2] = plt.subplots(1, 2, figsize=(15, 5)) 
# 建房时间的分布情况 
sns.distplot(df1['square'], bins=20, ax=ax1, color='r') 
sns.kdeplot(df1['square'], shade=True, ax=ax1) 
# 建房时间和出售价格的关系 
sns.regplot(x='square', y='price', data=df1, ax=ax2) 
plt.show()
```
